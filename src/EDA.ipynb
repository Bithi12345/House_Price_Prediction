```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('C:\\House_Price_Prediction\\data\\processed\\processed_data.csv')


# Set seaborn style
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# 1. Price Distribution
sns.histplot(df['Price'], bins=50, kde=True)
plt.title('Distribution of Property Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Area Distribution
sns.histplot(df['Area'], bins=50, kde=True, color='orange')
plt.title('Distribution of Area')
plt.xlabel('Area (sq ft)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. No. of Beds vs Price
sns.boxplot(x='No. Beds', y='Price', data=df)
plt.title('No. of Beds vs Price')
plt.xlabel('No. of Beds')
plt.ylabel('Price')
plt.show()

```

```python
# 4. No. of Baths vs Price
sns.boxplot(x='No. Baths', y='Price', data=df)
plt.title('No. of Baths vs Price')
plt.xlabel('No. of Baths')
plt.ylabel('Price')
plt.show()
```

```python
# 5. Price vs Area (scatter)
sns.scatterplot(x='Area', y='Price', data=df)
plt.title('Price vs Area')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()
```

```python

#6. Property Type vs Price
sns.boxplot(x='Type_n', y='Price', data=df)
plt.title('Property Type vs Price')
plt.xlabel('Type')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python

# 8. Correlation heatmap
corr = df[['Price', 'No. Beds', 'No. Baths', 'Area']].corr()
sns.heatmap(corr, annot=True, cmap='Blues', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 9. Jointplot of Area vs Price
sns.jointplot(data=df, x='Area', y='Price', kind='reg', height=7)
plt.show()
```

```python
# 10. Pairplot
sns.pairplot(df[['Price', 'No. Beds', 'No. Baths', 'Area']])
plt.show()
```
